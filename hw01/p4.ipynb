```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of N and M
N_values = np.arange(3, 101)  # N from 3 to 100
M_values = np.arange(2, 100)  # M from 2 to 99 (since M >= 2 and M < N)

# Create a meshgrid for N and M
N_grid, M_grid = np.meshgrid(N_values, M_values)

# Calculate P(EM,N)
# Ensure that M < N
valid_indices = M_grid < N_grid
P_EMN = np.zeros_like(N_grid, dtype=float)

# Compute P(EM,N) where valid
numerator = (N_grid - M_grid) * (N_grid - 1)
denominator = N_grid * (N_grid - 2)
P_EMN[valid_indices] = numerator[valid_indices] / denominator[valid_indices]

# Mask invalid values (where M >= N)
P_EMN[~valid_indices] = np.nan

# Plotting the heatmap
plt.figure(figsize=(12, 8))
plt.imshow(P_EMN, extent=[N_values.min(), N_values.max(), M_values.max(), M_values.min()],
           aspect='auto', cmap='viridis')

plt.colorbar(label='$P(E_{M,N})$')
plt.xlabel('Number of Doors (N)')
plt.ylabel('Number of Empty Doors (M)')
plt.title('Heatmap of $P(E_{M,N})$ for 3 ≤ N ≤ 100 and 2 ≤ M < N')

plt.show()
```
